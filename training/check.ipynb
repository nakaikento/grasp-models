```python
import os

def check_sync_quality(ja_path, ko_path, num_lines=100):
    # „Éï„Ç°„Ç§„É´„ÅÆÂ≠òÂú®Á¢∫Ë™ç
    if not os.path.exists(ja_path) or not os.path.exists(ko_path):
        print("‚ùå „Éï„Ç°„Ç§„É´„ÅåË¶ã„Å§„Åã„Çä„Åæ„Åõ„Çì„ÄÇ„Éë„Çπ„ÇíÁ¢∫Ë™ç„Åó„Å¶„Åè„Å†„Åï„ÅÑ„ÄÇ")
        return

    # Êó•Êú¨Ë™ûÔºàÂéüÊñáÔºâ„Å®ÈüìÂõΩË™ûÔºàÁøªË®≥Ê∏àÔºâ„ÇíË™≠„ÅøËæº„Åø
    with open(ja_path, 'r', encoding='utf-8') as f_ja:
        ja_lines = [line.strip() for line in f_ja]
    with open(ko_path, 'r', encoding='utf-8') as f_ko:
        ko_lines = [line.strip() for line in f_ko]

    current_ko_count = len(ko_lines)
    
    # --- Áµ±Ë®à„ÅÆË®àÁÆó ---
    # ÂÖ®‰Ωì„ÅÆFAILEDÊï∞„Çí„Ç´„Ç¶„É≥„Éà
    failed_total = sum(1 for line in ko_lines if "FAILED_TRANSLATION_CLEANED" in line)
    ok_total = current_ko_count - failed_total
    success_rate = (ok_total / current_ko_count * 100) if current_ko_count > 0 else 0

    print("=" * 50)
    print(f"üìä ÂÖ®‰Ωì„ÅÆÈÄ≤Êçó„ÉªÁµ±Ë®àÁä∂Ê≥Å")
    print(f"  Á∑èË°åÊï∞      : {len(ja_lines)} Ë°å")
    print(f"  ÂÆå‰∫ÜÊ∏à„Åø    : {current_ko_count} Ë°å (ÈÄ≤ÊçóÁéá: {current_ko_count / len(ja_lines) * 100:.2f}%)")
    print("-" * 50)
    print(f"  ‚úÖ ÊàêÂäü(OK) : {ok_total} Ë°å")
    print(f"  ‚ùå Â§±Êïó(FLD): {failed_total} Ë°å")
    print(f"  üìà Ê≠£Ëß£Áéá   : {success_rate:.2f} %")
    print("=" * 50 + "\n")

    # ÊúÄÊñ∞„ÅÆÁøªË®≥ÁµêÊûú„ÇíË°®Á§∫
    print(f"üîç ÊúÄÊñ∞„ÅÆÁøªË®≥„Éö„Ç¢ (Áõ¥Ëøë {num_lines}‰ª∂„ÅÆ„Çµ„É≥„Éó„É´)")
    start_point = max(0, current_ko_count - num_lines)
    
    for i in range(start_point, current_ko_count):
        ja = ja_lines[i]
        ko = ko_lines[i]
        
        status = "‚úÖ OK"
        if "FAILED" in ko:
            status = "‚ùå FAILED"
        elif len(ko) > 20 and ko.count(ko[:5]) > 2:
            status = "‚ö†Ô∏è LOOP?"

        print(f"[{i+1}Ë°åÁõÆ] {status}")
        print(f"  Êó•: {ja}")
        print(f"  Èüì: {ko}")
        print("-" * 30)

# „Éë„Çπ„ÇíÊåáÂÆö
JA_FILE = "mt-ja-ko/mt-ja-ko/data/raw/OpenSubtitles.ja-ko.ja"
KO_FILE = "mt-ja-ko/mt-ja-ko/data/teacher/train.ko"

# ÂÆüË°å
check_sync_quality(JA_FILE, KO_FILE)
```
